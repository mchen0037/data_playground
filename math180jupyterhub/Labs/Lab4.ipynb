```r
library(ISLR)
attach(Auto)
```

```r
set.seed(1)
trainingIndices=sample(392,196) #pick 196 random numbers out of 392 indices
## just for checks
length(trainingIndices)
```

```r
training=(Auto[trainingIndices,])
testing=(Auto[-trainingIndices,])
```

```r
model=lm(mpg~horsepower, data=training)
summary(model)
mean((mpg-predict(model, testing))^2) #this calculates the MSE.
```

```r
#Second model by using horsepower squared
model2 = lm(mpg~poly(horsepower, 2), data=training)
summary(model2)
mean((mpg-predict(model2, testing))^2) #this calculates the MSE.
```

```r
#Second model by using horsepower cubed
model3 = lm(mpg~poly(horsepower, 3), data=training)
summary(model3)
mean((mpg-predict(model3, testing))^2) #this calculates the MSE.
```

```r
#Change the training set and then recreate the mdoels
for (x in 2:10) {    
    set.seed(x)
    trainingIndices=sample(392,196)
    training=(Auto[trainingIndices,])
    testing=(Auto[-trainingIndices])
    model=lm(mpg~horsepower, data=training)
    model2 = lm(mpg~poly(horsepower, 2), data=training)
    model3 = lm(mpg~poly(horsepower, 3), data=training)
    a = mean((mpg-predict(model, testing))^2)
    b = mean((mpg-predict(model2, testing))^2) #this calculates the MSE.
    c = mean((mpg-predict(model3, testing))^2) #this calculates the MSE.
    print(paste0(a,"       ", b,"       ", c))

}
```
