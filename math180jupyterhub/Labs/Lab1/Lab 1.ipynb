### Histogram 1: Normal Distribution
Choose n numbers to generate histograms. Let's start small.

```r
n = 10
```

Using the following lines, we generate 10 random numbers that are normally distributed around 0 with a variance of 1.

```r
x = rnorm(n, mean=0, sd=1)
hist(x, xlab="Numbers", ylab="Frequency of Numbers", main="10 Random Generated Numbers", col="BLUE")
```

There isn't *too* much to see in this example, since we only sampled 10 numbers. Let's try again with a larger sample size.

```r
n = 100
x = rnorm(n, mean=0, sd=1)
hist(x, xlab="Numbers", ylab="Frequency of Numbers", main="100 Random Generated Numbers", col="BLUE")
```

We can see that it looks closer to the typical picture of a normal distribution! Let's scale up by another degree.

```r
n = 1000
x = rnorm(n, mean=0, sd=1)
hist(x, xlab="Numbers", ylab="Frequency of Numbers", main="1000 Random Generated Numbers", col="BLUE")
```

In this last example, we can see that this is a good representation of the uniform distribution! Yay! :)

### Histogram 2: Die Rolls

We want to create a histogram of random die rolls and how often we roll each number.

```r
minValue = 1;
maxValue = 6;
numtrials = 1000;
v = sample(minValue:maxValue,numtrials,TRUE);
hist(v, col='red', xlab="Die Value", ylab="Frequency of Die Value", main="1000 Random Die Rolls")
```

Each Die Value is pretty close, which is what we would expect from a fair die.
