This example protrays the idea that the number of rapes in each state (per 100,000 is not nearly in the same degree of magnitude as the number of people in urban population. We can use __principal component analysis__ to try to scale the variables together.

```r
pr.out = prcomp(USArrests, scale=TRUE)
```

```r
pr.out
names(pr.out)
```

```r
#this is the same as the mean that we printed out earlier.
pr.out$center
```

```r
pr.out$scale
```

```r
pr.out$rotation
```

```r
dim(USArrests)
dim(pr.out$x)
```

PCA Essentially generates a matrix to rotate the original data matrix by in order to explain the most variance with the least amount of variables.

```r
pr.out$x
```

```r
biplot(pr.out, scale=0)
```

```r
pr.out$rotation=-pr.out$rotation
pr.out$x=-pr.out$x
biplot(pr.out, scale=0)
```

```r
pr.out$sdev
```

```r
#sdev = sqrt(var)
pr.var=pr.out$sdev^2
pr.var
```

```r
#The proportion of variance explained by each prinicpal component = divide the variance explained by each PC
pve=pr.var/sum(pr.var)
pve
```

```r
plot(pve, xlab="PC", ,ylab="Proprtion of Variance Explained", ylim=c(0,1), type='b')
```

```r
plot(cumsum(pve), xlab="PC", ylab="cumulative proportion of variance explained", ylim=c(0,1), type='b')
```

Essentially, using the first two principal components, we can explain 62.0 + 24.7% of the variance. using all 4 PC's will give us 100% of the variance.

## 10.5: Clustering

```r
set.seed(2)
#Create DF with two columns of data
x=matrix(rnorm(50*2), ncol=2)
#Add 3 to first 25 of the x values of this DF and -4 to the y values.
x[1:25, 1] = x[1:25,1] + 3
x[1:25, 2] = x[1:25, 2] - 4
```

```r
x
```

```r
plot(x)
```

```r
# K=2 (i.e, two different classes.)
km.out=kmeans(x, 2, nstart=20)
km.out
names(km.out)
```

```r
#Centroids of the clusters--where the points are around.
km.out$centers
#The vector of the classes--you can use this to plot colors in plot()
km.out$cluster
```

```r
plot(x, col=(km.out$cluster+1), main="K-Means Clustering Results with K=2", pch=20, cex=2)
```

```r
set.seed(4)
km.out=kmeans(x, 3, nstart=1)
km.out
```

```r
plot(x, col=(km.out$cluster+1), 
     main="K-Means Clustering Results with K=3, nstart=1", pch=20, cex=2)
```

```r
km.out=kmeans(x, 3, nstart=20)
plot(x, col=(km.out$cluster+1), 
     main="K-Means Clustering Results with K=3, nstart=20", pch=20, cex=2)
```

## 10.5.2: Hierarchical Clustering

```r
x
```

```r
#The three different ways of using hierarchical clustering
hc.complete=hclust(dist(x), method="complete")
hc.average=hclust(dist(x), method="average")
hc.single=hclust(dist(x), method="single")
```

```r
plot(hc.complete, main="Complete Linkage")
```

```r
plot(hc.average, main="Average Linkage")
```

```r
plot(hc.single, main="Single Linkage")
```

```r
par(mfrow=c(1,3))
plot(hc.complete, main="Complete", cex=.9)
plot(hc.average, main="Average", cex=.9)
plot(hc.single, main="Single", cex=.9)
```

```r
cutree(hc.complete, 2)
```
