## Problem 3

```r
x = seq(from=0, to=1, by=.01)
```

```r
gini = 2* (x * (1 - x))
```

```r
class = list()
for (i in 0:length(x) - 1) {
    y = max(x[i], 1-x[i])
    class = append(class, 1 - y)
}
```

```r
ent = -1 * ( (x * log(x)) + ((1-x) * log(1-x)))
```

```r
plot(x,gini, col='red')
plot(x,class[2:102],col='blue')
plot(x,ent, col='green')
```
