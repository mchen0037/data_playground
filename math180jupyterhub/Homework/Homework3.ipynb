## Chapter 4, Problem 7

Normal Distribution takes the form $N\text{~ }(\mu,\sigma^2) = \frac{1}{\sqrt{2\pi}}e^{-\frac{1}{2}x^2}$. The two distributions that we have, "Issued a Dividend" (red) and "Did not Issue a Dividend" (blue) are graphed below and take the form $N\text{~ }(10,36)$ and $N\text{~ }(0,36)$, respectively.

```r
x <- seq(-20,30,length=1000)
y <- dnorm(x,mean=10, sd=6)
y2 <- dnorm(x, mean=0, sd=6)
plot(x,y, type="l", lwd=1, col='red')
lines(x,y2, type="l", lwd=1, col='blue')
```

We can use Baye's rule to estimate the probability.
$$ Pr(Y=yes|X=4) = \frac{0.8*N\text{~ }(10,36)_{x=4}}{0.8*N\text{~ }(10,36)_{x=4} + 0.2*N\text{~ }(0,36)_{x=4}}$$
$$ = \frac{0.8 * 0.0403}{0.8 * 0.0403 + 0.2 * 0.0532} = 75.2\%$$
and
$$ Pr(Y=no|X=4) = \frac{0.2*N\text{~ }(0,36)_{x=4}}{0.8*N\text{~ }(10,36)_{x=4} + 0.2*N\text{~ }(0,36)_{x=4}}$$
$$ = \frac{0.2* 0.0532}{0.8 * 0.0403 + 0.2 * 0.0532} = 24.8\%$$
