```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import re

%matplotlib inline
```

```python
df = pd.read_csv("data.csv")
```

```python
df.describe()
```

```python
for d in df.Date[0:5]:
    re.sub(r'[^a-zA-Z]', "", d)
```

```python
df['month'] = df['Date'].apply(lambda x: re.sub(r'[^a-zA-Z]', "", x))
```

## Total spent by month

```python
df.groupby(['month']).sum()
```

```python
df.loc[df['Type'] == 'Food'].groupby(['month']).sum()
```
